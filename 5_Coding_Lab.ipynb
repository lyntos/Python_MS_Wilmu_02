## Part 1: The Customer Class

    Create a Customer class with the following attributes:
        name: The name of the customer.
        address: The address of the customer.
    The Customer class should have the following methods:
        An initializer method that sets the name and address of the customer.
        Getter and setter methods for the name and address.

```python
# customer Class
class Customer:
  def __init__(self, name, address):
    self.name = name
    self.address = address

  def get_name(self):
    return self.name

  def set_name(self, name):
    self.name = name

  def get_address(self):
    return self.address

  def set_address(self, address):
    self.address = address

  def __str__(self):
    return f"Name: {self.name}\nAddress: {self.address}"
```

## Part 2: The Account Class

    Create an Account class with the following attributes:
        account_number: The account number.
        balance: The balance in the account.
        customer: A Customer object representing the account holder.
    The Account class should have the following methods:
        An initializer method that sets the account number, balance, and customer.
        Getter and setter methods for the account number, balance, and customer.
        deposit(amount): A method that increases the balance by the specified amount.
        withdraw(amount): A method that decreases the balance by the specified amount.



```python
class Account:
  def __init__(self, account_number, balance, customer):
    self.account_number = account_number
    self.balance = balance
    self.customer = customer

  def get_account_number(self):
    return self.account_number

  def set_account_number(self, account_number):
    self.account_number = account_number

  def get_balance(self):
    return self.balance

  def set_balance(self, balance):
    self.balance = balance

  def get_customer(self):
    return self.customer

  def set_customer(self, customer):
    self.customer = customer

  def deposit(self, amount):
    self.balance += amount
    return self.balance

  def withdraw(self, amount):
    self.balance -= amount
    return self.balance

  def __str__(self):
    return (f"Account Number: {self.account_number}\nBalance: {self.balance}\nCustomer: {self.customer}")
```

## Part 3: The Bank Class

    Create a Bank class with the following attributes:
        name: The name of the bank.
        accounts: A list of Account objects representing all accounts in the bank.
    The Bank class should have the following methods:
        An initializer method that sets the name of the bank and initializes the accounts list.
        add_account(account): A method that adds an Account object to the accounts list.
        get_account(account_number): A method that returns an Account object for the specified account number.

```python
class Bank:
  def __init__(self, name):
    self.name = name
    self.accounts = []

  def add_account(self, account):
    self.accounts.append(account)

  def get_account(self, account_number):
    for account in self.accounts:
      if account.account_number == account_number:
        return account
    return None

  def __str__(self):
    account_details = []
    for account in self.accounts:
      account_details.append(f"  Account Number: {account.get_account_number()} (Balance: {account.get_balance()})")
    return f"Bank Name: {self.name}\nAccounts:\n" + "\n".join(account_details) if account_details else f"Bank Name: {self.name}\nAccounts: None"
```
