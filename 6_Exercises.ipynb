> $base^{exponent} = base * base ^{exponent - 1}$

> and the terminating condition occurs when **exponent** is equal to 1, because

## 11.6 (Recursive Fibonacci Modification)

> Modify Section 11.4’s recursive **fibonacci** function to keep track of the total number of function calls. Display the number of calls for **fibonacci(10)**, **fibonacci(20)** and **fibonacci(30)**.

````python
def fibonacci(n):
  # instantiate calls for this function
  current_calls = 1

  if n in (0, 1):  # base cases
      return n, current_calls
  else:
      # recursively call and get the fib number and calls from sub-problems
      fib_n_minus_1, calls_n_minus_1 = fibonacci(n - 1)
      fib_n_minus_2, calls_n_minus_2 = fibonacci(n - 2)

      # sum the fib numbers and the calls
      total_fib = fib_n_minus_1 + fib_n_minus_2
      total_calls = current_calls + calls_n_minus_1 + calls_n_minus_2

      return total_fib, total_calls

fib_10, calls_10 = fibonacci(10)
print(f"fibonacci(10) = {fib_10:,}\t\t calls = {calls_10:,}")
%timeit -n 1 -r 10 fibonacci(10)

fib_20, calls_20 = fibonacci(20)
print(f"fibonacci(20) = {fib_20:,}\t\t calls = {calls_20:,}")
%timeit -n 1 -r 10 fibonacci(20)

fib_30, calls_30 = fibonacci(30)
print(f"fibonacci(30) = {fib_30:,}\t\t calls = {calls_30:,}")
%timeit -n 1 -r 10 fibonacci(30)

fib_35, calls_35  = fibonacci(35)
print(f"fibonacci(35) = {fib_35:,}\t calls = {calls_35:,}")
%timeit -n 1 -r 10 fibonacci(35)
````

## 11.7 (Improving Recursive Fibonacci Performance: Memoization)

> Research the performance enhancement technique called memoization. Modify Section 11.4’s recursive  function to incorporate memoization. Compare the performance of both versions for **fibonacci(10)**, **fibonacci(20)** and **fibonacci(30)**.

````python
# memoized recursive Fibonacci sequence
import numpy as np

def memoized_fibonacci_wrapper(n):
    # initialize memoization table for each top-level call
    fib_memo = np.zeros(n + 1, dtype=int) # Use int for fib numbers

    def _fib_recursive(k):
        if fib_memo[k] != 0: # Check if already computed
            return fib_memo[k]
        if k <= 1:
            fib_memo[k] = k
            return k
        else:
            result = _fib_recursive(k - 1) + _fib_recursive(k - 2)
            fib_memo[k] = result
            return result
    return _fib_recursive(n)

fib_memoized_10 = memoized_fibonacci_wrapper(10)
print(f"fibonacci(10) = {fib_memoized_10:,}")
%timeit -n 1 -r 1000 memoized_fibonacci_wrapper(10)

fib_memoized_20 = memoized_fibonacci_wrapper(20)
print(f"fibonacci(20) = {fib_memoized_20:,}")
%timeit -n 1 -r 1000 memoized_fibonacci_wrapper(20)

fib_memoized_30 = memoized_fibonacci_wrapper(30)
print(f"fibonacci(30) = {fib_memoized_30:,}")
%timeit -n 1 -r 1000 memoized_fibonacci_wrapper(30)

fib_memoized_35 = memoized_fibonacci_wrapper(35)
print(f"fibonacci(35) = {fib_memoized_35:,}")
%timeit -n 1 -r 1000 memoized_fibonacci_wrapper(35)
````

## 11.10 (Palindromes)

> A palindrome is a string that is spelled the same way forward and backward. Some examples of palindromes are “radar,” “able was i ere i saw elba” and (if spaces are ignored) “a man a plan a canal panama.”

> Write a recursive **test_palindrome** function that returns **True** if the string stored in an array is a palindrome and  otherwise. The function should ignore spaces and punctuation in the string.

````python
def test_palindrome(string):
  a_array = list(string.lower())
  b_array = a_array.copy()
  b_array.reverse()
  if a_array == b_array:
    return True
````

````python
def test_palindrome_recursive(s):
  '''recursively check for palindrome by comparing s[0] and s[-1]'''
  # convert to lowercase and remove non-alphanumeric characters
  cleaned_s = "".join(char for char in s.lower() if char.isalnum())

  # Base case - empty string or a single-character string is a palindrome
  if len(cleaned_s) <= 1:
    return True
  # recursive step - check if the first and last characters match,
  # then recursively call the function on the substring between them
  elif cleaned_s[0] == cleaned_s[-1]:
    return test_palindrome_recursive(cleaned_s[1:-1])
  # if characters don't match, it's not a palindrome
  else:
    return False

# Test cases
print(f"'radar' is a palindrome: {test_palindrome_recursive('radar')}")
print(f"'able was i ere i saw elba' is a palindrome: {test_palindrome_recursive('able was i ere i saw elba')}")
print(f"'a man a plan a canal panama' is a palindrome: {test_palindrome_recursive('a man a plan a canal panama')}")
print(f"'hello' is a palindrome: {test_palindrome_recursive('hello')}")
````

## 11.11 (Eight Queens)

> A puzzler for chess buffs is the Eight Queens problem, which asks: Is it possible to place eight queens on an empty chessboard so that no queen is “attacking” any other (i.e., no two queens are in the same row, in the same column or along the same diagonal)? For example, if a queen is placed in the upper-left corner of the board, no other queens could be placed in any of the marked squares shown in the following figure. Solve the problem recursively.

> [Hint: Your solution should begin with the first column and look for a location in that column where a queen can be placed—initially, place the queen in the first row. The solution should then recursively search the remaining columns. In the first few columns, there will be several locations where a queen may be placed. Take the first available location. If a column is reached with no possible location for a queen, the program should return to the previous column, and move the queen in that column to a new row. This continuous backing up and trying new alternatives is an example of recursive backtracking.]


````python
def is_safe(board, row, col):
  # check this row on left side
  for i in range(col):
    if board[row][i] == 1:
      return False

  # check upper diagonal on left side
  for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
    if board[i][j] == 1:
      return False

  # check lower diagonal on left side
  for i, j in zip(range(row, 8, 1), range(col, -1, -1)):
    if board[i][j] == 1:
      return False

  return True

def solve_eight_queens(board, col):
  # base case: if all queens are placed, return true
  if col >= 8:
    return True

  # consider this column and try placing this queen in all rows one by one
  for row in range(8):
    if is_safe(board, row, col):
      # place this queen in board[row][col]
      board[row][col] = 1

      # recur to place the rest of the queens
      if solve_eight_queens(board, col + 1):
        return True

      # if placing queen in board[row][col] doesn't lead to a solution,
      # then remove queen from board[row][col] (backtrack)
      board[row][col] = 0

  # if the queen cannot be placed in any row in this column, return false
  return False

def print_solution(board, show_threats=False):
    for row in board:
      print(" ".join("Q" if x == 1 else "." for x in row))

# initialize the 8x8 chessboard with all zeros
board = [[0 for _ in range(8)] for _ in range(8)]

if solve_eight_queens(board, 0):
  print("solution for eight queens problem:")
  print_solution(board)
else:
  print("solution does not exist")
````

      1 -> 3
      2 -> 1
      2 -> 3
      1 -> 3

````python
def towers_of_hanoi(n, source_peg, destination_peg, auxiliary_peg):
  # base case: if only one disk, move it directly
  if n == 1:
    print(f"{source_peg} -> {destination_peg}")
    return

  # move n-1 disks from source to auxiliary, using destination as temp
  towers_of_hanoi(n - 1, source_peg, auxiliary_peg, destination_peg)

  # move the largest disk from source to destination
  print(f"{source_peg} -> {destination_peg}")

  # move n-1 disks from auxiliary to destination, using source as temp
  towers_of_hanoi(n - 1, auxiliary_peg, destination_peg, source_peg)

print("Moves for 3 disks from peg 1 to peg 3:")
towers_of_hanoi(3, 1, 3, 2)

print("\nMoves for 4 disks from peg 1 to peg 3:")
towers_of_hanoi(4, 1, 3, 2)

print("\nMoves for 5 disks from peg 1 to peg 3:")
towers_of_hanoi(5, 1, 3, 2)
````

## 11.17 Bucket Sort

> A bucket sort begins with a one-dimensional array of positive integers to be sorted and a two-dimensional array of integers with rows indexed from 0 to 9 and columns indexed from 0 to $n -1$, where n is the number of values to be sorted. Each row of the two-dimensional array is referred to as a bucket. Write a class named **BucketSort** containing a function called **sort** that operates as follows:
              
    a. Place each value of the one-dimensional array into a row of the bucket array, based on the value’s “ones” (rightmost) digit.
    For example, 97 is placed in row 7, 3 is placed in row 3, and 100 is placed in row 0.
    This procedure is called a distribution pass.
                
    b. Loop through the bucket array row by row, and copy the values back to the original array.
    This procedure is called a gathering pass.
    The new order of the preceding values in the one-dimensional array is 100, 3 and 97.
                
    c. Repeat this process for each subsequent digit position (tens, hundreds, thousands, etc.).
    On the second (tens digit) pass, 100 is placed in row 0, 3 is placed in row 0 (because 3 has no tens digit), and 97 is placed in row 9.
    After the gathering pass, the order of the values in the one-dimensional array is 100, 3 and 97.
    On the third (hundreds digit) pass, 100 is placed in row 1, 3 is placed in row 0, and 97 is placed in row 0 (after the 3).
    After this last gathering pass, the original array is in sorted order.
                
> The two-dimensional array of buckets is 10 times the length of the integer array being sorted. This sorting technique provides better performance than a selection and insertion sorts but requires much more memory—the selection and insertion sorts require space for only one additional element of data. This comparison is an example of a space/time trade-off: The bucket sort uses more memory than the selection and insertion sorts, but performs better. This version of the bucket sort requires copying all the data back to the original array on each pass. Another possibility is to create a second two-dimensional bucket array and repeatedly swap the data between the two bucket arrays.

````python
class BucketSort:
  def __init__(self, data):
    self.data = list(data)

  def get_digit(self, num, digit_position):
    return (num // 10**digit_position) % 10

  def sort(self):
    # determine the maximum number of digits in the array
    max_val = 0
    if self.data:
      max_val = max(self.data)

    max_digits = 0
    if max_val > 0:
      max_digits = len(str(max_val))
    elif self.data and max_val == 0:
      max_digits = 1
    elif not self.data:
        return self.data

    # iterate through each digit position (ones, tens, hundreds, etc.)
    for digit_position in range(max_digits):
      # create 10 buckets
      buckets = [[] for _ in range(10)]

      # distribution pass: place each number into the appropriate bucket
      for number in self.data:
        digit = self.get_digit(number, digit_position)
        buckets[digit].append(number)

      # gathering pass: copy numbers from buckets back to the original array
      self.data.clear() # Clear the original array
      for bucket in buckets:
        for number in bucket:
          self.data.append(number)

    return self.data # return the sorted list

# test an array
arr1 = [35, 73, 90, 65, 23, 86, 43]
bucket_sort = BucketSort(arr1)
sorted_arr1 = bucket_sort.sort()
print(sorted_arr1)
%timeit -n 1 -r 10 bucket_sort.sort()
````

## 11.18 (Recursive Binary Search)

> Modify this chapter’s **binary_search** function to perform a recursive binary search of the array. The function should receive the search key, starting index and ending index as arguments. If the search key is found, return its index in the array. If the search key is not found, return $-1$.

````python
import numpy as np

def recursive_binary_search(data, key, start, end):
    if start > end:
        return -1

    middle = (start + end) // 2

    # if the key is found at the middle
    if key == data[middle]:
        return middle
    # if the key is in the left half
    elif key < data[middle]:
        return recursive_binary_search(data, key, start, middle - 1)
    # if the key is in the right half
    else:
        return recursive_binary_search(data, key, middle + 1, end)


sorted_data = np.array([2, 5, 8, 12, 16, 23, 38, 56, 72, 91])
print(f"Sorted array: {sorted_data}")

# test cases
key_to_find = 23
index = recursive_binary_search(sorted_data, key_to_find, 0, len(sorted_data) - 1)
print(f"Key {key_to_find} found at index: {index}" if index != -1 else f"Key {key_to_find} not found")


key_to_find = 91
index = recursive_binary_search(sorted_data, key_to_find, 0, len(sorted_data) - 1)
print(f"Key {key_to_find} found at index: {index}" if index != -1 else f"Key {key_to_find} not found")
````

## 11.19 (Quicksort)

> The recursive sorting technique called quicksort uses the following basic algorithm for a one-dimensional array of values:
              
                
    a. Partitioning Step: Take the first element of the unsorted array and determine its final location in the sorted array
    (i.e., all values to the left of the element in the array are less than the element,
    and all values to the right of the element in the array are greater than the element
    —we show how to do this below).
    We now have one element in its proper location and two unsorted subarrays.
                
    b. Recursive Step: Perform Step 1 on each unsorted subarray.
    Each time Step 1 is performed on a subarray, another element is placed in its final location of the sorted array, and two unsorted subarrays are created.
    When a subarray consists of one element, that element is in its final location (because a one-element array is already sorted).
    The basic algorithm seems simple enough, but how do we determine the final position of the first element of each subarray?
    As an example, consider the following set of values (the element in bold is the partitioning element—it will be placed in its final location in the sorted array):

>       37 2 6 4 89 8 10 12 68 45           

> The basic algorithm seems simple enough, but how do we determine the final position of the first element of each subarray? As an example, consider the following set of values (the element in bold is the partitioning element—it will be placed in its final location in the sorted array):

> Starting from the rightmost element of the array, compare each element with 37 until an element less than 37 is found; then swap 37 and that element. The first element less than 37 is 12, so 37 and 12 are swapped. The new array is

>       *12* 2 6 4 89 8 10 37 68 45

Element 12 is in "* *" to indicate that it was just swapped with 37.
                
> Starting from the left of the array, but beginning with the element after 12, compare each element with 37 until an element greater than 37 is found—then swap 37 and that element. The first element greater than 37 is 89, so 37 and 89 are swapped. The new array is

>       12 2 6 4 37 8 10 *89* 68 45

> Starting from the right, but beginning with the element before 89, compare each element with 37 until an element less than 37 is found—then swap 37 and that element. The first element less than 37 is 10, so 37 and 10 are swapped. The new array is

>       12 2 6 4 *10* 8 37 89 68 45

> Starting from the left, but beginning with the element after 10, compare each element with 37 until an element greater than 37 is found—then swap 37 and that element. There are no more elements greater than 37, so when we compare 37 with itself, we know that 37 has been placed in its final location in the sorted array. Every value to the left of 37 is smaller than it, and every value to the right of 37 is larger than it.
                
> Once the partition has been applied on the previous array, there are two unsorted subarrays. The subarray with values less than 37 contains 12, 2, 6, 4, 10 and 8. The subarray with values greater than 37 contains 89, 68 and 45. The sort continues recursively, with both subarrays being partitioned in the same manner as the original array.
                
Based on the preceding discussion, write recursive function **quick_sort_helper** to sort a one-dimensional integer array. The function should receive as arguments a starting index and an ending index on the original array being sorted. Call this function from a **quick_sort** function that receives just the original array to sort.


````python
def quick_sort(data):
  """Initiates the recursive quicksort process."""
  # call the helper function to perform the recursive sort
  if data:
    quick_sort_helper(data, 0, len(data) - 1)

def partition(data, left, right):
  pivot_index = left # choose the first element as the pivot
  pivot_value = data[pivot_index]
  left_pointer = left + 1 # start from the element after the pivot
  right_pointer = right # start from the last element

  done = False
  while not done:
    # move left_pointer to the right until an element > pivot_value is found
    while left_pointer <= right_pointer and data[left_pointer] <= pivot_value:
      left_pointer += 1

    # move right_pointer to the left until an element < pivot_value is found
    while data[right_pointer] >= pivot_value and right_pointer >= left_pointer:
      right_pointer -= 1

    # if right_pointer has crossed left_pointer, then we are done partitioning
    if right_pointer < left_pointer:
      done = True
    else:
      # swap elements at left_pointer and right_pointer
      data[left_pointer], data[right_pointer] = data[right_pointer], data[left_pointer]

  # swap the pivot element with the element at right_pointer's position
  # this places the pivot in its final sorted position
  data[pivot_index], data[right_pointer] = data[right_pointer], data[pivot_index]

  return right_pointer # return the final position of the pivot

def quick_sort_helper(data, first, last):
  if first < last:
    # partition the array and get the pivot's final index
    split_point = partition(data, first, last)

    # recursively sort the two subarrays
    quick_sort_helper(data, first, split_point - 1)
    quick_sort_helper(data, split_point + 1, last)


arr1 = [37, 2, 6, 4, 89, 8, 10, 12, 68, 45]
print(f"Original array: {arr1}")
quick_sort(arr1)
print(f"Sorted array:   {arr1}\n")
````
