##1. Email Extraction:
> Write a Python program to extract all email addresses from a given string. The string can be a paragraph of text or a webpage content



```python
# extract emails from a test paragraph
import re
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}'

test_paragraph = r"The project kicked off with an update from dev.lead_99@internal-system.net, though the server logs flagged an error from admin-checker@status.io shortly after. If you run into issues, please don't reach out to totally_fake_person@not-a-real-domain.org as that mailbox is unmonitored. Instead, forward your diagnostic reports to beta_testing_group_v2@sandbox.cloud-service.com or the secondary alias user.name+extra-data@prototype.dev. For billing inquiries, you can look up the records for billing-dept@corp-structure.co.uk to see if they align with the automated pings from bot-01@automation.farm."

emails = []

for email in re.findall(email_pattern, test_paragraph):
  emails.append(email)

print(emails)
```

##2. URL Extraction:
> Write a Python program to find all URLs in a string. The string can be a paragraph of text or a webpage content.


```python
import re

# this truncates and returns the core URL
pattern = r'https?://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}'

# this includes the full link
full_pattern = r'https?://(?:www\.)?[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}(?:/[a-zA-Z0-9.-]+)*'

# test paragraph with valid and invalid URLs
test_string = "Please review the following links: https://www.example.com and http://blog.site.co.uk/page?id=123 are valid, as is https://sub.domain.org/path/file.html. However, be careful with invalid ones like hhtp:/wrong.com, www.missing-protocol.com, and https://space in path.com. Also, check ftp://valid-protocol.net, but ignore invalid://missing.dot and https://example.com/trailing/space."

# apply regex for core and full URLs
core_urls = re.findall(pattern, test_string)
full_urls = re.findall(full_pattern, test_string)

print(core_urls)
print(full_urls)

```

```python
import re

#using 10-digit phone numbers only. a country code lookup and format would make this more robust

# create pattern for any non-digit characters
clean_pattern = r'\D'

# create pattern for 10 digits
length_pattern = r'^\d{10}$'

test_phone_numbers = ['(123) 456-7891', '1234567', '808-456-7892', '302.456.7893', '1234567894', '(865)-46-7895', '123456785000']

# initialize list to store munged numbers
formatted_numbers = []

# loop through phone numbers, apply clean_pattern and length_pattern
for number in test_phone_numbers:
  number = re.sub(clean_pattern, '', number)
  if re.match(length_pattern, number):
    formatted_numbers.append(f'({number[:3]}){number[3:6]}-{number[6:]}')
  # apply a standard phone number format once cleaned

print(formatted_numbers)
```

##4. Password Strength Check:
> Write a Python program to check the strength of a password. The password should contain at least one uppercase letter, one lowercase letter, one special character, one digit, and have a minimum length of 8 characters.


```python
import re

def calculate_password_strength_score(password):
  '''score the password based on given criteria'''
  score = 0

  # length points based on number of characters
  score += len(password)

  # check presence of character types
  has_uppercase = re.search(r'[A-Z]', password)
  has_lowercase = re.search(r'[a-z]', password)
  has_digit = re.search(r'\d', password)
  has_special = re.search(r'[!@#$%<^>&*?]', password)

  char_types_present = 0
  if has_uppercase:
    score += 3
    char_types_present += 1
  if has_lowercase:
    score += 3
    char_types_present += 1
  if has_digit:
    score += 3
    char_types_present += 1
  if has_special:
    score += 3
    char_types_present += 1

  # bonus for diversity
  if char_types_present >= 3:
    score += 4
  if char_types_present == 4:
    score += 2

  # dictionary word check (downgrade score) - limited word list for demonstration
  common_dictionary_words = [
      "password", "123456", "qwerty", "admin", "strong", "user", "welcome", "secret",
      "test", "dragon", "love", "computer", "qweasd", "football", "soccer"
  ]

  password_lower = password.lower()

  for word in common_dictionary_words:
      if word in password_lower:
          # downgrade score by 1 point for each character in the matched dictionary word
          score -= len(word)*4 # penalize for ease of hacking
      if score <= 0:
        score = round(len(password)/3)
        break

  return score

def get_password_strength_grade(score):
  if score <= 10:
    return 'Weakest'
  elif score <= 15:
    return 'Weaker'
  elif score <= 20:
    return 'Weak'
  elif score <= 25:
    return 'Okay'
  elif score <= 30:
    return 'Good'
  elif score <= 35:
    return 'Strong'
  elif score <= 40:
    return 'Stronger'
  else:
    return 'Strongest'

# Test cases
test_passwords = {
    1: "pass",
    2: "password",
    3: "Password",
    4: "Password#123",
    5: "My_45Stro_ngP@_ss2w0_!rd!",
    6: "ThisIsAsecretVeryLongAndStrongadminPassword123!@#",
    7: "1234567890",
    8: "abcdefghij",
    9: "MyPassword123!",
    10: "user123456",
    11: "_NoD1ct!W0rd@Here_",
    12: "lE04{7*T\\£96c(w2+S2q*TJG.D(3q", # generated from LastPass, https://www.lastpass.com/features/password-generator
    13: "My45StrongP@ss2w0rd!"
    }

print("Password Strength Results:")
for i, pwd in enumerate(test_passwords.values()):
  score = calculate_password_strength_score(pwd)
  grade = get_password_strength_grade(score)
  print(f"Password: '{pwd:<}'\n  Score: {score:>3}\t  Grade: {grade:<}\n")
```

##6. HTML Tag Removal:

```python
import re
# all html tags start with < and end with > - remove those and everything in between
html_pattern = r'<[^>]+>' # start with tag opening(<), get 1 or more of every character but tag closing(>), then look for tag closing (>)

html_string = '<div class="container" id="main-content"><h1 style="color: blue;">Welcome to the <span class="highlight">Regex Testing Ground</span></h1><p class="intro-text">This is a sample paragraph with a <a href="https://example.com" target="_blank">hidden link</a> and some <strong>bold text</strong>.</p><ul id="data-list"><li class="item-1">First item with <i class="icon"></i> metadata.</li><li class="item-2">Second item with a <br /> self-closing tag.</li></ul><!-- This is a comment that should also be removed --><script type="text/javascript">console.log("This script content should be handled carefully.");</script><footer data-info={"version": "1.0.4"}>&copy; 2024 Practice Corp.</footer></div>"'

html_stripped = re.sub(html_pattern, '\n', html_string)

print(html_stripped)
```

##8. Credit Card Number Validation:
> Write a Python program to validate credit card numbers. The credit card number should be 16 digits long and pass the Luhn check.


```python
# luhn algorithm
# https://en.wikipedia.org/wiki/Luhn_algorithm
def luhn_checksum(card_number):
  # drop the last digit
  digits = digits_of(card_number)
  odd_digits = digits[-1::-2]
  even_digits = digits[-2::-2]
  checksum = 0
  checksum += sum(odd_digits)
  for digit in even_digits:
    checksum += sum(digits_of(digit * 2))
  return checksum % 10

def digits_of(number):
  return [int(d) for d in str(number)]

def check_length(card_number):
  return len(card_number) == 16

# these are random 16-digit numbers
test_credit_cards = ['1234565809423459', '4111111111111111', '5435657812341544']

for card in test_credit_cards:
  digits = print(luhn_checksum(card))
  if luhn_checksum(card) == 0 and check_length(card):
    print(f'{card} is a valid credit card number')
  else:
    print(f'{card} is an invalid credit card number')

```

##10. Word Frequency Count:
> Write a Python program to count the frequency of each word in a text. Ignore common stop words such as a, an, the, in, is, it, etc

```python
import re

# word frequency counter
text = r"""The quick, brown fox jumps over 12 lazy dogs—or so the well-known legend says. Don't forget that it's 2024, and "state-of-the-art" technology is everywhere! If we travel to the U.S.A., will we see more foxes? Maybe. 100% of the researchers say, "Yes!" However, cost-effective solutions are rare. It's a bit tricky... right?"""

# list stop words
stop_words = ['a', 'an', 'the', 'in','is', 'of', 'or', 'am', 'be', 'if', 'then', 'to']

# add each word to a dictionary and count occurrences
word_count = {}
for word in re.findall(r"(\w+\w+'*\w*)", text.lower()):
  if word in stop_words: # this is a quick fix, we could also drop them from the dictionary
    word_count[word] = 0
  elif word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1

# print neatly
for word, count in sorted(word_count.items()):
  if count != 0:
    print(f'{word}: {count}')
```
