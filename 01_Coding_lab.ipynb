```python
#ask user to input a number, determine if it is +/-
userinput = input('Enter a number: ')
userinput=int(userinput)
if(isinstance(userinput, int)):
  print('You have entered ', userinput)
  if isinstance(userinput, int):
    if userinput > 0:
      print(userinput,' is postive')
    elif userinput == 0:
      print (userinput,' is zero')
    else:
      print (userinput,' is negative')
else:
  print ('Please enter a number');
```

```python

#ask user for input, determine odd/even
userinput = input('Enter a number: ')
userinput=int(userinput)

if(isinstance(userinput, int)):
  print('You have entered', userinput)
  if isinstance(userinput, int):
    if userinput % 2 == 0:
      print(userinput,'is even')
    else:
      print (userinput,'is odd')
else:
  print ('Please enter a number');
```

```python
#user inputs number n, program sums from 1 to n
userinput = input('Enter a number: ')
userinput=int(userinput)
output = 0
if(isinstance(userinput, int)):
  print('You have entered ', userinput)
  for i in range(1,userinput+1):
    output += i
  print (f'The sum of 1 through {userinput} is {output}')
else:
  print ('Please enter a number');
```

```python
#user inputs number n, program prints multiplication table from 1 to n
userinput = input('Enter a number: ')
userinput=int(userinput)
output = 0
if(isinstance(userinput, int)):
  print('You have entered ', userinput)
  for i in range(1,11):
    output = userinput * i
    print (f'{userinput} X {i} = {output}')
else:
  print ('Please enter a number');
```
