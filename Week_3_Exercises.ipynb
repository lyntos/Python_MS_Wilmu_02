##7.19 (Enhanced Median and Mode of an array)
Modify your functions from the previous exercise to allow the user to provide an axis keyword argument so the calculations can be performed row-by-row or column-by-column on a two-dimensional array.

````python
# create functions for row-wise and column-wise median and mean of an array
def row_median(array, x):
  '''calculate the median of each row'''
  if x in (0,1):
  # x is row or column
    return np.median(array, axis = x)
  else:
    return False

def row_mode(array, x):
  '''calculate the mode of each row/column based on axis'''
  if x in (0,1):
  # x is row or column
    if x == 1:
      # row wise mode
      return np.apply_along_axis(lambda x: np.bincount(x).argmax(), 1, array)
    if x == 0:
      # column wise mode
      return np.apply_along_axis(lambda x: np.bincount(x).argmax(), 0, array)
  else:
    return False

# print outputs using all three arrays and 0,1,2 as args
print("Axis: Column")
print(f"Median: {row_median(array1, 1)}")
print(f"Mode: {row_mode(array1, 1)}")
print(f"Median: {row_median(array2, 1)}")
print(f"Mode: {row_mode(array2, 1)}")
print(f"Median: {row_median(array3, 1)}")
print(f"Mode: {row_mode(array3, 1)}")

print("Axis: Row")
print(f"Median: {row_median(array1, 0)}")
print(f"Mode: {row_mode(array1, 0)}")
print(f"Median: {row_median(array2, 0)}")
print(f"Mode: {row_mode(array2, 0)}")
print(f"Median: {row_median(array3, 0)}")
print(f"Mode: {row_mode(array3, 0)}")

print("Axis: Invalid")
print(f"Median: {row_median(array1, 2)}")
print(f"Mode: {row_mode(array1, 2)}")
print(f"Median: {row_median(array2, 2)}")
print(f"Mode: {row_mode(array2, 2)}")
print(f"Median: {row_median(array3, 2)}")
print(f"Mode: {row_mode(array3, 2)}")

# to make this more robust, we should go back and add *args
# this way, if the user does not pass an axis, it will return median and mode
# for the entire array
````

> c. Select from temperatures the column of temperature readings for 'Maxine'.
              
> d. Select from temperatures the row of 'Morning' temperature readings.
              
> e. Select from temperatures the rows for 'Morning' and 'Evening' temperature readings.
              
> f. Select from temperatures the columns of temperature readings for 'Amanda' and 'Maxine'.
              
> g. Select from temperatures the elements for 'Amanda' and 'Maxine' in the 'Morning' and 'Evening'.
              
> h. Use the describe method to produce temperatures’ descriptive statistics.
              
> i. Transpose temperatures.
              
> j. Sort temperatures so that its column names are in alphabetical order.

````python
import pandas as pd
import numpy as np

# a. Create a DataFrame named temperatures from a dictionary of three temperature readings each for 'Maxine', 'James' and 'Amanda'.
# values are random 0.1f between 97.0 and 104.0
dictionary_temps = {'Maxine': np.random.uniform(97.0, 104.0,3), 'James': np.random.uniform(97.0, 104.0,3), 'Amanda': np.random.uniform(97.0, 104.0,3)}
temperatures = pd.DataFrame(dictionary_temps)
print(temperatures)
# b. Recreate the DataFrame temperatures in Part (a) with custom indices using the  keyword
# argument and a list containing 'Morning', 'Afternoon' and 'Evening'.

dictionary_temps_time = {'Maxine': np.random.uniform(97.0, 104.0,3), 'James': np.random.uniform(97.0, 104.0,3), 'Amanda': np.random.uniform(97.0, 104.0,3)}
temperatures = pd.DataFrame((dictionary_temps_time), index = ['Morning', 'Afternoon', 'Evening'])
print(temperatures)

# c. Select from temperatures the column of temperature readings for 'Maxine'.
# print column for maxine
print(temperatures[['Maxine']])

# d. Select from temperatures the row of 'Morning' temperature readings.
# print row 'Morning'
print(temperatures.loc[['Morning']])

# e. Select from temperatures the rows for 'Morning' and 'Evening' temperature readings.
print(temperatures.loc[['Morning', 'Evening']])

# f. Select from temperatures the columns of temperature readings for 'Amanda' and 'Maxine'.
print(temperatures[['Amanda', 'Maxine']])

# g. Select from temperatures the elements for 'Amanda' and 'Maxine' in the 'Morning' and 'Evening'.
print(temperatures.loc[['Morning','Evening'],['Amanda','Maxine']])
# h. Use the describe method to produce temperatures’ descriptive statistics.

# i. Transpose temperatures.
print(temperatures.transpose())

# j. Sort temperatures so that its column names are in alphabetical order.
print(temperatures.sort_index(axis = 1))
````
