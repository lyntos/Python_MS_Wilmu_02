### 8.1 (Check Protection)
Although electronic deposit has become extremely popular, payroll and accounts payable applications often print checks. A serious problem is the intentional alteration of a check amount by someone who plans to cash a check fraudulently. To prevent a dollar amount from being altered, some computerized check-writing systems employ a technique called check protection. Checks designed for printing by computer typically contain a fixed number of spaces for the printed amount. Suppose a paycheck contains eight blank spaces in which the computer is supposed to print the amount of a weekly paycheck. If the amount is large, then all eight of the spaces will be filled:

Check Amount | Position Numbers
-----|-----
1,230.60 | 01234567

On the other hand, if the amount is smaller, then several of the spaces would ordinarily be left blank. For example, contains two blank spaces. If a check is printed with blank spaces, it’s easier for someone to alter the amount. Check-writing systems often insert leading asterisks to prevent alteration and protect the amount as follows:

Check Amount | Position Numbers
-----|-----
**399.87 | 01234567

Write a script that inputs a dollar amount, then prints the amount in check-protected format in a field of 10 characters with leading asterisks if necessary. [Hint: In a format string that explicitly specifies alignment with <, ^ or >, you can precede the alignment specifier with the fill character of your choice.]

```python
def format_check(check_string):
  # check if there is a decimal entered
  insert_decimal = False
  if '.' in check_string:
    insert_decimal = True

  check_string = check_string.replace("$", "").replace(",", "").replace(".", "")
  check_string = check_string.zfill(10)
  check_string = check_string.lstrip('0')

  print(check_string)

  # convert the string to a list
  check_list = list(check_string)

  # insert a '.' in position -2 if there was a decimal entered earlier
  if insert_decimal:
    check_list.insert(-2, '.')
  else:
    check_list.append('.')
    check_list.append('0')
    check_list.append('0')

  # get length of check_list
  check_len = len(check_list)

  # add leading asterisks to the list until the list in 10 characters long
  while check_len < 10:
      check_list.insert(0, '*')
      check_len = len(check_list)

  #print(check_list)

  # join each element of the list back into a string
  check_string = ''.join(check_list)
  return(check_string)


check1 = format_check(input("Enter a check amount: "))
print(check1)

check2 = format_check(input("Enter a check amount: "))
print(check2)
```

Create a sentence by selecting a word at random from each array in the following order: [article], [noun], [verb], [preposition], [article] and [noun]. As each word is picked, concatenate it to the previous words in the sentence. Spaces should separate the words. When the final sentence is output, it should start with a capital letter and end with a period. The script should generate and display 20 sentences.

```python
import numpy as np

# initialize string placeholder arrays
nouns = np.full(10,'', dtype= object)
articles = np.full(3, '' ,dtype= object)
verbs = np.full(10, '',dtype = object)
prepositions = np.full(10, '', dtype = object)

noun_list = ['person','place','idea','name','thought']
article_list = ['a','the', 'an']
verb_list = ['do','know','be','act','give','take','enjoy']
preposition_list = ['above','below','from','to','of','behind','around','under','over']

# fill the arrays with random words from the lists
for i in range(10):
  nouns[i] = np.random.choice(noun_list)
  verbs[i] = np.random.choice(verb_list)
  prepositions[i] = np.random.choice(preposition_list)

for i in range(3):
  articles[i] = article_list[i]

# initialize sentence variable
sentence = ''

# create sentences in the order article, noun, verb, preposition, article, noun
for i in range(20):
  sentence = np.random.choice(articles).capitalize()
  sentence += ' '
  sentence += np.random.choice(nouns)
  sentence += ' '
  sentence += np.random.choice(verbs)
  sentence += ' '
  sentence += np.random.choice(prepositions)
  sentence += ' '
  sentence += np.random.choice(articles)
  sentence += ' '
  sentence += np.random.choice(nouns)
  sentence += '.'
  print(sentence)
  sentence = ''

```

### 8.11 (Project: Evaluate Word Problems)
Write a script that enables the user to enter mathematical word problems like “two times three” and “seven minus five”, then use string processing to break apart the string into the numbers and the operation and return the result. So “two times three” would return 6 and “seven minus five” would return 2. To keep things simple, assume the user enters only the words for the numbers 0 through 9 and only the operations 'plus', 'minus', 'times' and 'divided by'.

```python
import numpy as np

operands = {'plus': '+', 'minus':'-', 'times':'*', 'divided':'/'}
numbers = {'zero':0, 'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9}

math_string = input('Enter an arithmetic word problem with numbers 0-9:')

components = math_string.split()

# find the operand, get the value from the key in operands{}
for key, value in operands.items():
  if key in components:
    operand = value

# get the values from numbers{} as provided by components
numeric_values = []
for key, value in numbers.items():
  if key in components:
    numeric_values.append(value)

# switch order for division
if operand != '/':
  print(f"Expression: {numeric_values[0]} {operand} {numeric_values[1]}")
else:
  print(f"Expression: {numeric_values[1]} {operand} {numeric_values[0]}")

result = 0
if operand == '+':
  result = numeric_values[0] + numeric_values[1]
elif operand == '-':
  result = numeric_values[0] - numeric_values[1]
elif operand == '*':
  result = numeric_values[0] * numeric_values[1]
elif operand == '/':
  if numeric_values[0] != 0:
    result = numeric_values[1] / numeric_values[0]
  else:
    print("Error: Division by zero")

print(f"Result: {result}")
```

### 8.12 (Project: Scrambled Text)
Use string-processing capabilities to keep the first and last letter of a word and scramble the remaining letters in between the first and last. Search online for “University of Cambridge scrambled text” for an intriguing paper on the readability of texts consisting of such scrambled words. Investigate the random module’s shuffle function to help you implement this exercise’s solution.

```python
'''Scrambled text'''
import numpy as np # Ensure numpy is imported for np.random.shuffle
import re # Ensure re is imported for re.sub

# Assuming tokenize function from previous cells is available (e.g., DmDyN-tXlPYG)
def tokenize(string):
  '''split the word into tokens'''
  # strip non-letter and non-whitespace chars
  phrase = re.sub(r'[^​\w\s]','',string)
  tokens = phrase.split()
  return tokens

phrase = 'A bunch of words that need to be mixed up'

# remove all punctuation from phrase using re

# assign tokens to an array
tokens = tokenize(phrase)

# create and object array to store scrambled words
scrambled_words = np.array([], dtype = object)

for token in tokens:
  if len(token) < 3:
    # Words with less than 3 letters cannot have their 'middle' scrambled
    new_token = token
  else:
    first_letter = token[0]
    last_letter = token[-1]
    # Extract middle letters
    middle_letters = list(token[1:-1])
    # Shuffle the middle letters
    np.random.shuffle(middle_letters)
    # Reconstruct the word
    new_token = first_letter + ''.join(middle_letters) + last_letter

  scrambled_words = np.append(scrambled_words, new_token)

new_phrase = ' '.join(scrambled_words)
print(new_phrase)
```

### 8.16 (Regular Expressions: Locating URLs)

### 8.17 (Regular Expressions: Matching Numeric Values)
Write a regular expression that searches a string and matches a valid number. A number can have any number of digits, but it can have only digits and a decimal point and possibly a leading sign. The decimal point is optional, but if it appears in the number, there must be only one, and it must have digits on its left and its right. There should be whitespace or a beginning or end-of-line character on either side of a valid number.
              
                

```python
import re
# https://docs.python.org/3/library/re.html#regular-expression-examples

# pattern 1 - optional + or - [+-]?
# pattern 2 - any number of digits \d+ (one or more digits)
# pattern 3 - can contain one decimal, but must have digits on both sides (\.?)
# pattern 4 - check for trailing digits

def check_valid_number(string):
  check_valid_number = re.compile(r'[-+]?\d+\.?\d+?')
  length_check = check_valid_number.findall(str(string))
  if len(length_check) == 1:
    return length_check
  else:
    return False

valid_number = '100022.65'
invalid_number = '1.00101.1'

print(check_valid_number(valid_number))
print(check_valid_number(invalid_number))
```

### 8.18 (Regular Expression: Password Format Validator)
Search online for secure password recommendations, then research existing regular expressions that validate secure passwords. Two examples of password requirements are:
> • Passwords must contain at least five words, each separated by a hyphen, a space, a period, a comma or an underscore.
              
> • Passwords must have a minimum of 8 characters and contain at least one each from uppercase characters, lowercase characters, digits and punctuation characters (such as characters in '!@#$%<^>&*?').

Write regular expressions for each of the two requirements above, then use them to test sample passwords.


```python
import re

# must contain at least five words
# each separated by a hyphen, space, period, comma or underscore
def password_requirement_one(password):
  # delimiters: hyphen, space, period, comma, underscore
  # match one or more (+)
  delimiter_pattern = r'[-_\s.,]+'

  # split the password by these delimiters
  words = re.split(delimiter_pattern, password)

  # filter out empty strings
  actual_words = [word for word in words if word]

  return len(actual_words) >= 5 # returns true if 5 or more words are found

# minimum of 8 characters and contain at least one each from
# uppercase characters, lowercase characters, digits and punctuation characters '!@#$%<^>&*?'
def password_requirement_two(password):
  # minimum 8 characters
  if len(password) < 8:
    return False

  # at least one uppercase character
  if not re.search(r'[A-Z]', password): # use search to find anywhere in string
    return False

  # at least one lowercase character
  if not re.search(r'[a-z]', password): # use search to find anywhere in string
    return False

  # at least one digit
  if not re.search(r'\d', password): # use search to find anywhere in string
    return False

  # at least one punctuation character from the specified examples
  punctuation_pattern = r'[!@#$%<^>&*?]'
  if not re.search(punctuation_pattern, password):
    return False
  return True

# Test cases
password1 = 'space-hyphen_underscore.period,comma' # long enough
password2 = 'SpACe-HYph3N_UnD3R$c0R3,+comma- PERIOD!' # good password
password3 = 'Apple-styl3-p@$ww0Rd_but longer?' # needs more words

print(f"Password 1 ('{password1}'):")
print(f"\tRequirement 1 met: {password_requirement_one(password1)}")
print(f"\tRequirement 2 met: {password_requirement_two(password1)}")

print(f"\nPassword 2 ('{password2}'):")
print(f"\tRequirement 1 met: {password_requirement_one(password2)}")
print(f"\tRequirement 2 met: {password_requirement_two(password2)}")

print(f"\nPassword 3 ('{password3}'):")
print(f"\tRequirement 1 met: {password_requirement_one(password3)}")
print(f"\tRequirement 2 met: {password_requirement_two(password3)}")
```

```python
import re

date_containing_string = r"""ENTRY_ID:101 | DATE:2023-10-14 | STATUS:OK; ID:102 | DATE:12/05/2022 | STATUS:PENDING; USER:j_doe | LOGIN:01-Jan-2024 | LOC:NY; TS:1704067200 | EVENT:LOGOUT | UID:44; DATE:2021.08.19 14:30:05 | FLAG:TRUE; DATE:Mar 15, 2023 | REF:9920; UPDATE:22/02/24 | VER:2.1; DATE:2025/01/01 | NOTE:New Year; TIMESTAMP: November 16, 2022 | USER:admin; DATE:05-12-21 | TYPE:LEGACY """

# create a pattern of date several date patterns ... | == OR
date_pattern = r'(\b\d{4}[-/.]\d{2}[-/.]\d{2}\b|\b\d{1,2}[-/.]\d{1,2}[-/.]\d{2,4}\b|\b\d{6}\b|\b\d{1,2}[-/]\w{3}[-/]\d{2,4}\b|\b\w+\s+\d{1,2}(?:,\s*)?\s+\d{2,4}\b)'

dates_found = re.findall(date_pattern, date_containing_string)

# Reverse lookup for month names to numbers
month_name_to_num = {
    'jan': '01', 'feb': '02', 'mar': '03', 'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09', 'oct': '10', 'nov': '11', 'dec': '12',
    'january': '01', 'february': '02', 'march': '03', 'april': '04', 'june': '06',
    'july': '07', 'august': '08', 'september': '09', 'october': '10', 'november': '11', 'december': '12'
}

print("Original dates found:")
for date_str in dates_found:
  print(date_str)

print("\nDates reformatted to MM/DD/YYYY:")

for date_str in dates_found:
  month, day, year = None, None, None

  # YYYY-MM-DD or YYYY.MM.DD or YYYY/MM/DD
  if re.match(r'\b\d{4}[-/.]\d{2}[-/.]\d{2}\b', date_str):
    year_p, month_p, day_p = re.split('[-/.]', date_str)
    month, day, year = month_p.zfill(2), day_p.zfill(2), year_p
  # MM/DD/YYYY or DD/MM/YYYY or MM-DD-YY
  # asserting MM/DD/YYYY or MM/DD/YY based on common US convention - this could be adjusted if we knew the person/procedure for entering data. Without SOP, context, or locale, we can't know for sure
  elif re.match(r'\b\d{1,2}[-/.]\d{1,2}[-/.]\d{2,4}\b', date_str):
    parts = re.split('[-/.]', date_str)
    # check length of parts to differentiate between MM/DD/YYYY and MM/DD/YY
    if len(parts[2]) == 4: # MM/DD/YYYY
      month_p, day_p, year_p = parts[0], parts[1], parts[2]
    else: # MM/DD/YY
      month_p, day_p, year_p = parts[0], parts[1], parts[2]
      # heuristic for 2-digit year to 4-digit year
      year_p = '20' + year_p if int(year_p) < 50 else '19' + year_p
    month, day, year = month_p.zfill(2), day_p.zfill(2), year_p # use zfill to standardize the date 'width' with leading zeroes
  # MMDDYY format
  elif re.match(r'\b\d{6}\b', date_str):
    month_p, day_p, year_p = date_str[0:2], date_str[2:4], date_str[4:6]
    year_p = '20' + year_p if int(year_p) < 50 else '19' + year_p
    month, day, year = month_p.zfill(2), day_p.zfill(2), year_p
  # DD-Mon-YYYY format (e.g., 01-Jan-2024)
  elif re.match(r'\b\d{1,2}[-/]\w{3}[-/]\d{2,4}\b', date_str, re.IGNORECASE):
    parts = re.split('[-/]', date_str)
    day_p = parts[0]
    month_p = month_name_to_num.get(parts[1].lower(), None)
    year_p = parts[2]
    if month_p:
      month, day, year = month_p.zfill(2), day_p.zfill(2), year_p
  # Month Day, YYYY or Month Day YYYY (e.g., Mar 15, 2023, November 16, 2022)
  elif (month_day_year_match := re.match(r'(\w+)\s+(\d{1,2})(?:,\s*)?\s+(\d{2,4})', date_str, re.IGNORECASE)):
    month_p = month_name_to_num.get(month_day_year_match.group(1).lower(), None)
    day_p = month_day_year_match.group(2)
    year_p = month_day_year_match.group(3)
    if month_p:
      month, day, year = month_p.zfill(2), day_p.zfill(2), year_p

  if all([month, day, year]):
    print(f"{month}/{day}/{year}")
  else:
    print(f"Could not re-format '{date_str}'")
```

 8.21(Project: Metric Conversions) Write a script that assists the user with some common metric-to-English conversions. Your script should allow the user to specify the names of the units as strings (i.e., centimeters, liters, grams, and so on for the metric system and inches, quarts, pounds, and so on for the English system) and should respond to simple questions, such as

      'How many inches are in 2 meters?'
      'How many litres are in 10 quarts?'

> Your script should recognize invalid conversions. For example, the following question is not meaningful, because 'feet' is a unit of length and 'kilogram' is a unit of mass:

      'How many feet are in 5 kilograms?'

> Assume that all questions are in the form shown above. Use regular expressions to capture the important substrings, such as 'inches' , '2' and 'meters' in the first sample question above. Recall that functions int and float can convert strings to numbers.

```python
%pip install nltk
import re
import nltk
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

# create a conversion dictionary with original keys
raw_unit_data = {
    # Distance (base: meter)
    'meter': (1.0, 'distance'),
    'inches': (0.0254, 'distance'),
    'feet': (0.3048, 'distance'),
    'yards': (0.9144, 'distance'),
    'miles': (1609.34, 'distance'),
    'centimeters': (0.01, 'distance'),
    'kilometers': (1000.0, 'distance'),

    # Volume (base: liter)
    'liter': (1.0, 'volume'),
    'cups': (0.236588, 'volume'),
    'quarts': (0.946353, 'volume'),
    'gallons': (3.78541, 'volume'),
    'fluid ounces': (0.0295735, 'volume'),
    'milliliters': (0.001, 'volume'),

    # Mass (base: kilogram)
    'kilogram': (1.0, 'mass'),
    'kilograms': (1.0, 'mass'),
    'pounds': (0.453592, 'mass'),
    'ounces': (0.0283495, 'mass'),
    'grams': (0.001, 'mass'),
    'milligrams': (0.000001, 'mass'),
    'metric tons': (1000.0, 'mass'),
    'tonnes': (1000.0, 'mass')
}

# create a new unit_data dictionary with stemmed keys
unit_data = {}
for key, value in raw_unit_data.items():
    # stem each word in multi-word units and join them back
    stemmed_key_parts = [stemmer.stem(word) for word in key.split()]
    stemmed_key = ' '.join(stemmed_key_parts)
    unit_data[stemmed_key] = value

# get user input for conversion type
conversion_type_input = input('What would you like to convert? (distance (d), volume (v), or mass (m)): \n').lower()

# filter unit_data based on user input
filtered_unit_data = {}
if conversion_type_input == 'd':
  filtered_unit_data = {k: v for k, v in unit_data.items() if v[1] == 'distance'}
  print("Available distance units:")
  for k in filtered_unit_data.keys():
    print(f"{k:<12}")
elif conversion_type_input == 'v':
  filtered_unit_data = {k: v for k, v in unit_data.items() if v[1] == 'volume'}
  print("Available volume units:")
  for k in filtered_unit_data.keys():
    print(f"{k:<12}")
elif conversion_type_input == 'm':
  filtered_unit_data = {k: v for k, v in unit_data.items() if v[1] == 'mass'}
  print("Available mass units:")
  for k in filtered_unit_data.keys():
    print(f"{k:<12}")
else:
  print("Invalid conversion type selected.")


if filtered_unit_data:
  question = input("\nEnter your conversion question (e.g., 'How many inches are in 2 meters?'):\n")

  # regex to capture target_unit, value, and source_unit
  pattern = r'How many (\w+(?:\s\w+)*) are in ([\d\.]+) (\w+(?:\s\w+)*)(\?)*'
  match = re.search(pattern, question, re.IGNORECASE)

  if match:
    target_unit_raw = match.group(1).lower()
    value = float(match.group(2))
    source_unit_raw = match.group(3).lower()

    # stem the units extracted from the user's question
    target_unit_parts = [stemmer.stem(word) for word in target_unit_raw.split()]
    target_unit = ' '.join(target_unit_parts)

    source_unit_parts = [stemmer.stem(word) for word in source_unit_raw.split()]
    source_unit = ' '.join(source_unit_parts)

    # Check if units are in the filtered_unit_data (which now has stemmed keys)
    if target_unit in filtered_unit_data and source_unit in filtered_unit_data:
      # Check if unit types are compatible
      if filtered_unit_data[target_unit][1] == filtered_unit_data[source_unit][1]:
        # Perform conversion
        value_in_base = value * filtered_unit_data[source_unit][0]
        converted_value = value_in_base / filtered_unit_data[target_unit][0]

        print(f"\n{question} Result: {converted_value:.6e} {target_unit_raw}")
      else:
        print("\nInvalid conversion: Units are not of the same type.")
    else:
      print("\nInvalid conversion: One or both units not recognized or not available for the selected type (after stemming).")
  else:
    print("\nInvalid question format. Please use 'How many [unit1] are in [number] [unit2]?'")
else:
  print("No conversion possible due to invalid conversion type selection.")
```
