## Part a: Brute-Force Knight's Tour - Single Random Tour

> Use random-number generation to enable the knight to walk around the chessboard (in its legitimate L-shaped moves) at random. Your script should run one tour and display the final chessboard. How far did the knight get?


```python
import numpy as np
import random

def knights_tour_brute_force_single():
    '''run the brute force knight's tour'''
    #define L-shaped moves (provided by textook)
    horizontal = [2, 1, -1, -2, -2, -1, 1, 2]
    vertical = [1, 2, 2, 1, -1, -2, -2, -1]

    # initialize 8x8 chessboard with all squares marked as unvisited (0)
    board_size = 8
    chessboard = np.zeros((board_size, board_size))

    # initialize random starting position
    current_row = random.randint(0, board_size - 1) # this ensures the knight starts on a legal row square
    current_col = random.randint(0, board_size - 1) # this ensures the knight starts on a legal column square

    # place knight at random starting position
    # Mark it as visited with '1' (first move)
    move_count = 1
    chessboard[current_row, current_col] = move_count

    print(f"Starting position: ({current_row}, {current_col})")

    # loop for the tour
    while True:
        # initialize empty list to store moves made
        possible_moves = []
        for i in range(len(horizontal)):
            #brute force square selection based on horizontal and vertical array combinations of i
            next_row = current_row + vertical[i]
            next_col = current_col + horizontal[i]

            # check if the move is within board boundaries and to an unvisited square
            if (0 <= next_row < board_size and
                    0 <= next_col < board_size and
                    chessboard[next_row, next_col] == 0):
                possible_moves.append((next_row, next_col)) # add legal moves to the possible_moves array

        # if there are no legal moves, break the loop
        if not possible_moves:
            break

        # assign the position from a random choice from the possible_moves array
        chosen_move = random.choice(possible_moves)
        current_row, current_col = chosen_move

        # move the knight to the chosen new position and count the move
        move_count += 1
        chessboard[current_row, current_col] = move_count

    # print the final state of the chessboard
    print("\nFinal Chessboard:")
    for row in chessboard:
        print(' '.join(f'{cell:2}' for cell in row))

    # print the total number of squares the knight visited
    print(f"\nTotal squares visited: {move_count}")
    return move_count, chessboard

# run the tour
knights_tour_brute_force_single()
```

## Part b: Brute-Force Knight's Tour - 1,000,000 Tours Analysis

Modify the script from Part (a) to perform 1,000,000 random Knight's Tours. Use an array to store the count of tours for each length, and finally display this information in a neat tabular format to identify the best result.


```python
import numpy as np
import pandas as pd
import random

def run_single_tour():
    # Define the possible L-shaped moves
    horizontal = [2, 1, -1, -2, -2, -1, 1, 2]
    vertical = [1, 2, 2, 1, -1, -2, -2, -1]

    board_size = 8
    chessboard = np.zeros((board_size, board_size), dtype=int)

    # Choose a random starting position
    current_row = random.randint(0, board_size - 1)
    current_col = random.randint(0, board_size - 1)

    move_count = 1
    chessboard[current_row, current_col] = move_count

    while True:
        possible_moves = []
        for i in range(len(horizontal)):
            next_row = current_row + vertical[i]
            next_col = current_col + horizontal[i]

            # Check if the move is within board boundaries and to an unvisited square
            if (0 <= next_row < board_size and
                    0 <= next_col < board_size and
                    chessboard[next_row, next_col] == 0):
                possible_moves.append((next_row, next_col))

        if not possible_moves:
            break

        chosen_move = random.choice(possible_moves)
        current_row, current_col = chosen_move

        move_count += 1
        chessboard[current_row, current_col] = move_count

    return move_count

def knights_tour_brute_force_multiple():
  '''run the brute force knight's tour multiple times'''
    #run 1M tours
    num_tours = 1_000_000
    tour_lengths_counts = np.zeros(65) # Max tour length is 64

    print(f"Running {num_tours} random Knight's Tours...")

    for _ in range(num_tours):
        length = run_single_tour()
        tour_lengths_counts[length] += 1

    print("\n--- Tour Length Distribution ---")
    # initialize empty list to store results
    results = []
    # set max tour length to 0 - this will be updated later
    max_length_achieved = 0
    # loop through
    for length, count in enumerate(tour_lengths_counts):
        if count > 0:
            results.append({'Tour Length': length, 'Number of Tours': count})
            # update max tour length achevied
            if length > max_length_achieved:
                max_length_achieved = length

    df_results = pd.DataFrame(results)
    print(df_results.to_string(index=False))

    print(f"\nMaximum tour length achieved: {max_length_achieved}")

# Execute the multiple tour simulation
knights_tour_brute_force_multiple()
```

## Part c: Brute-Force Knight's Tour - First Full Tour

> c. Most likely, the script in Part (b) gave you some “respectable” tours, but no full tours. Now let your script run until it produces a full tour. [Caution: This version of the script could run for hours on a powerful computer.] Once again, keep a table of the number of tours of each length, and display this table when the first full tour is found. How many tours did your script attempt before producing a full tour? How much time did it take?

```python
import numpy as np
import random
import pandas as pd
import time

def run_single_tour_with_board():
    horizontal = [2, 1, -1, -2, -2, -1, 1, 2]
    vertical = [1, 2, 2, 1, -1, -2, -2, -1]

    board_size = 8
    chessboard = np.zeros((board_size, board_size), dtype=int)

    current_row = random.randint(0, board_size - 1)
    current_col = random.randint(0, board_size - 1)

    move_count = 1
    chessboard[current_row, current_col] = move_count

    while True:
        possible_moves = []
        for i in range(len(horizontal)):
            next_row = current_row + vertical[i]
            next_col = current_col + horizontal[i]

            if (0 <= next_row < board_size and
                    0 <= next_col < board_size and
                    chessboard[next_row, next_col] == 0):
                possible_moves.append((next_row, next_col))

        if not possible_moves:
            break

        chosen_move = random.choice(possible_moves)
        current_row, current_col = chosen_move

        move_count += 1
        chessboard[current_row, current_col] = move_count

    return move_count, chessboard

def knights_tour_brute_force_full_tour():
    attempts_count = 0
    full_tour_found = False
    tour_lengths_counts = np.zeros(65, dtype=int) # Max tour length is 64
    full_tour_board = None

    print("Searching for the first full Knight's Tour (64 squares)...")
    start_time = time.time()

    while not full_tour_found:
        attempts_count += 1
        length, board = run_single_tour_with_board()
        tour_lengths_counts[length] += 1

        if length == 64:
            full_tour_found = True
            full_tour_board = board
            end_time = time.time()
            break

        # Optional: print progress every N attempts
        if attempts_count % 1_000_000 == 0:
            print(f"Attempt: {attempts_count:,}. Max length so far: {np.max(np.where(tour_lengths_counts > 0))}")

    time_taken = end_time - start_time

    print(f"\n--- Full Tour Found! ---")
    print(f"Total attempts: {attempts_count:,}")
    print(f"Time taken: {time_taken:.2f} seconds")

    print("\n--- Tour Length Distribution at time of full tour discovery ---")
    results = []
    for length, count in enumerate(tour_lengths_counts):
        if count > 0:
            results.append({'Tour Length': length, 'Number of Tours': count})

    df_results = pd.DataFrame(results)
    print(df_results.to_string(index=False))

    print("\n--- Full Tour Chessboard ---")
    if full_tour_board is not None:
        for row in full_tour_board:
            print(' '.join(f'{cell:2}' for cell in row))
    else:
        print("Error: Full tour board not recorded.")

# Execute the full tour search
knights_tour_brute_force_full_tour()
```
